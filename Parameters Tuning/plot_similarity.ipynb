```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sim11 = np.load('similarity11.npy')
sim12 = np.load('similarity12.npy')
sim13 = np.load('similarity13.npy')
sim21 = np.load('similarity21.npy')
sim22 = np.load('similarity22.npy')
sim23 = np.load('similarity23.npy')

epochs = np.linspace(0,399,400)

plt.figure()
g1, = plt.plot(epochs,sim11,ls='-',lw=1,color = 'lightcoral')
g2, = plt.plot(epochs,sim12,ls='-',lw=1,color = 'y')
g3, = plt.plot(epochs,sim13,ls='-',lw=1,color = 'skyblue')
g4, = plt.plot(epochs,sim21,ls='-',lw=1,color = 'darkorange')
g5, = plt.plot(epochs,sim22,ls='-',lw=1,color = 'violet')
g6, = plt.plot(epochs,sim23,ls='-',lw=1,color = 'mediumseagreen')


plt.xlabel('400 epoch')
plt.ylabel('simmilarity')


plt.legend(handles=[g1,g2,g3,g4,g5,g6],labels=['1L1F','1L2F','1L3F','2L1F','2L2F','2L3F'],bbox_to_anchor=(1.12,0),loc=4,borderaxespad=0,fontsize=6)

plt.savefig('6_similarity_dwq.png',dpi=600)

plt.show()
```
