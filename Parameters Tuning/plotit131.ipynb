```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#画loss图像
x = np.linspace(0,399,400)
y = np.load('loss131_400epochs.npy')
plt.plot(x,y,ls='-',lw=1,label='loss',color = 'orange')
plt.legend()
plt.ylabel('loss')
plt.xlabel('400 epoch')

plt.savefig('loss_with_epoch.png',dpi=600)



plt.show()
```

```python
#画准确率图像
x = np.linspace(0,399,400)
y = np.load('acc131_400epoch.npy')   
plt.plot(x,y,ls='-',lw=1,label='accuracy_400epoch',color = 'orange')
plt.legend()
plt.ylabel('accuracy')
plt.xlabel('400 epoch')
plt.savefig('accuracy_with_epoch.png',dpi=600)

plt.show()
```

```python
#画平均相似度图像
x = np.linspace(0,399,400)
y = np.load('probablity_400epoch.npy')
plt.plot(x,y,ls='-',lw=1,label='average_similarity',color = 'orange')
plt.legend()
plt.ylabel('average similarity')
plt.xlabel('400 epoch')
plt.savefig('average_similarity_with_epoch.png',dpi=600)
plt.show()
```

```python
#画平均不相似度图像
x = np.linspace(0,399,400)
y = np.load('probablityng_400epoch.npy')
plt.plot(x,y,ls='-',lw=1,label='dissimilarity',color = 'orange')
plt.legend()
plt.ylabel('dissimilarity')
plt.xlabel('400 epoch')

plt.savefig('dissimilarity_with_epoch.png',dpi=600)
plt.show()
```

```python
#画损失和准确率图像
fig,ax1 = plt.subplots()
t = np.linspace(0,399,400)
s1 = np.load('loss131_400epochs.npy')
l1, = ax1.plot(t,s1,color='blue',ls='-',lw=1,label='loss')



ax1.set_xlabel('400 epochs')
ax1.set_ylabel('loss range with epoch')

ax2 = ax1.twinx()
s2 = np.load('acc131_400epoch.npy')
l2, = ax2.plot(t,s2,color='orange',ls='-',lw=1,label='accuracy')
ax2.set_ylabel('accuracy range with epoch')

plt.legend(handles=[l1,l2],labels=['loss','accuracy'],loc=5)
plt.savefig('loss_and_accuracy_with_epoch.png',dpi=600)
plt.show()
```

```python
#画损失和平均相似度的图像

fig,ax1 = plt.subplots()
t = np.linspace(0,399,400)
s1 = np.load('loss131_400epochs.npy')
l1, = ax1.plot(t,s1,color='blue',ls='-',lw=1)

ax1.set_xlabel('400 epochs')
ax1.set_ylabel('loss range with epoch')

ax2 = ax1.twinx()
s2 = np.load('probablity_400epoch.npy')
l2, = ax2.plot(t,s2,color='orange',ls='-',lw=1)
ax2.set_ylabel('simiraty range with epoch')

plt.legend(handles=[l1,l2],labels=['loss','similarity'],loc=5)
plt.savefig('loss_and_similarity_with_epoch.png',dpi=600)
plt.show()
```

```python
#画损失和平均不相似度的图像
fig,ax1 = plt.subplots()
t = np.linspace(0,399,400)
s1 = np.load('loss131_400epochs.npy')
l1, = ax1.plot(t,s1,color='blue',ls='-',lw=1)

ax1.set_xlabel('400 epochs')
ax1.set_ylabel('loss range with epoch')
ax1.tick_params('y',color='red')

ax2 = ax1.twinx()
s2 = np.load('probablityng_400epoch.npy')
l2, = ax2.plot(t,s2,color='orange',ls='-',lw=1)
ax2.set_ylabel('dissimilarity range with epoch')

plt.legend(handles=[l1,l2],labels=['loss','dissimilarity'],loc=1)
plt.savefig('loss_and_dissimilarity_with_epoch.png',dpi=600)
plt.show()
```

```python
#画平均相似度和平均不相似度的图像
fig,ax1 = plt.subplots()
t = np.linspace(0,399,400)
s1 = np.load('probablity_400epoch.npy')
l1, = ax1.plot(t,s1,color='darkcyan',ls='-',lw=1)

ax1.set_xlabel('400 epochs')
ax1.set_ylabel('similarity with epoch')

ax2 = ax1.twinx()
s2 = np.load('probablityng_400epoch.npy')
l2, = ax2.plot(t,s2,color='chocolate',ls='-',lw=1)
ax2.set_ylabel('dissimilarity range with epoch')

plt.legend(handles=[l1,l2],labels=['similarity','dissimilarity'],loc=5)
plt.savefig('similarity_and_dissimilarity_with_epoch.png',dpi=600)
plt.show()
```
