```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
loss11 = np.load('loss11.npy')
loss12 = np.load('loss12.npy')
loss13 = np.load('loss13.npy')
loss21 = np.load('loss21.npy')
loss22 = np.load('loss22.npy')
loss23 = np.load('loss23.npy')

epochs = np.linspace(0,399,400)

plt.figure()
g1, = plt.plot(epochs,loss11,ls='-',lw=1,color = 'lightcoral')
g2, = plt.plot(epochs,loss12,ls='-',lw=1,color = 'y')
g3, = plt.plot(epochs,loss13,ls='-',lw=1,color = 'skyblue')
g4, = plt.plot(epochs,loss21,ls='-',lw=1,color = 'darkorange')
g5, = plt.plot(epochs,loss22,ls='-',lw=1,color = 'violet')
g6, = plt.plot(epochs,loss23,ls='-',lw=1,color = 'mediumseagreen')


plt.xlabel('400 epoch')
plt.ylabel('loss')

plt.legend(handles=[g1,g2,g3,g4,g5,g6],labels=['1L1F','1L2F','1L3F','2L1F','2L2F','2L3F'],bbox_to_anchor=(1.12,0),loc=4,borderaxespad=0,fontsize=6)

plt.savefig('6_loss_dwq.png',dpi=600)

plt.show()
```
