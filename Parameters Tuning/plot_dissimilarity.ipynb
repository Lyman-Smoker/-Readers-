```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dis11 = np.load('dissimilarity11.npy')
dis12 = np.load('dissimilarity12.npy')
dis13 = np.load('dissimilarity13.npy')
dis21 = np.load('dissimilarity21.npy')
dis22 = np.load('dissimilarity22.npy')
dis23 = np.load('dissimilarity23.npy')

epochs = np.linspace(0,399,400)

plt.figure()
g1, = plt.plot(epochs,dis11,ls='-',lw=1,color = 'lightcoral')
g2, = plt.plot(epochs,dis12,ls='-',lw=1,color = 'y')
g3, = plt.plot(epochs,dis13,ls='-',lw=1,color = 'skyblue')
g4, = plt.plot(epochs,dis21,ls='-',lw=1,color = 'darkorange')
g5, = plt.plot(epochs,dis22,ls='-',lw=1,color = 'violet')
g6, = plt.plot(epochs,dis23,ls='-',lw=1,color = 'mediumseagreen')


plt.xlabel('400 epoch')
plt.ylabel('dissimilarity')

plt.legend(handles=[g1,g2,g3,g4,g5,g6],labels=['1L1F','1L2F','1L3F','2L1F','2L2F','2L3F'],bbox_to_anchor=(1.12,0),loc=4,borderaxespad=0,fontsize=6)

plt.savefig('6_dissimilarity_dwq.png',dpi=600)

plt.show()
```
