```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
acc11 = np.load('accuracy11.npy')
acc12 = np.load('accuracy12.npy')
acc13 = np.load('accuracy13.npy')
acc21 = np.load('accuracy21.npy')
acc22 = np.load('accuracy22.npy')
acc23 = np.load('accuracy23.npy')

epochs = np.linspace(0,399,400)

plt.figure()
g1, = plt.plot(epochs,acc11,ls='-',lw=1,color = 'lightcoral')
g2, = plt.plot(epochs,acc12,ls='-',lw=1,color = 'Y')
g3, = plt.plot(epochs,acc13,ls='-',lw=1,color = 'skyblue')
g4, = plt.plot(epochs,acc21,ls='-',lw=1,color = 'darkorange')
g5, = plt.plot(epochs,acc22,ls='-',lw=1,color = 'violet')
g6, = plt.plot(epochs,acc23,ls='-',lw=1,color = 'mediumseagreen')


plt.xlabel('400 epoch')
plt.ylabel('accuracy')

plt.legend(handles=[g1,g2,g3,g4,g5,g6],labels=['1L1F','1L2F','1L3F','2L1F','2L2F','2L3F'],bbox_to_anchor=(1.12,0),loc=4,borderaxespad=0,fontsize=6)

plt.savefig('6_accuracy_dwq.png',dpi=600)

plt.show()
```
